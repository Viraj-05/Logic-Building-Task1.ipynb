# Search query key word

```python
# for this task we first have to import string 

import string

search_query  = input("Enter your search query: ")
search_query = search_query.lower()
search_query = search_query.translate(str.maketrans('', '',string.punctuation))
words = search_query.split()

freqq = {}
for word in words  :
    freqq [word] = freqq.get(word, 0) + 1

repeated_keywords = {word: count  for word, count in freqq.items() if count > 1 }

print(repeated_keywords)
```

# Email Domain usage Analysis

```python
emails = ["alok@gmail.com", "vaibhav@outlook.com", "vijay@yahoo.com", "aadi@outlook.com", "vira@gmail.com"]
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_emails = len(emails)
print("Email Domain Usage: ")
for domain, count in domain_count.items():
    percentage = (count /total_emails) * 100
    print(f"{domain}:{percentage:.0f}%")
```
