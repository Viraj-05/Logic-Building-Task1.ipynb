#Error Message Detector

```python
#In this task we only have to count the number of "ERROR" message

logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count = 0


for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total errror count: ",error_count)
```

#Message Length Analyzer

```python
#Here we have to Print the length of each message and flag messages longer than 10 characters

messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages :
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")

    if length > 10:
        print("!!!10 characters exceeded!!!")
        
```

#Pass or fail Analyzer

```python
#for completing this task we have to first Count the total number of pass students and count the total number of fail students just like academic evaluation system

marks = [45, 78 , 90 , 33 , 60]
pass_students = 0
fail_students = 0
  
for mark in marks:
 if mark>= 50:
  pass_students += 1
 else :
  fail_students += 1

print("Passed Students :", pass_students)
print("Failed Students:", fail_students)
```

#Simple data Cleaneer

```python
#In this task we are given names with uneven spacing , poor formatting ,etc. and we have to make it clean

names = [" Alice ", "bob", " CHARLIE "]
clean_names = []

for name in names:
    formatted_names = name.strip().lower()
    clean_names.append(formatted_names)

print("Clean Names:", clean_names)


```
