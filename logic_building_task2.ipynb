# Unique Words Identifier

```python
sentence = input("Enter your Sentence: ")
sentence = sentence.lower()
words_in_sentence = sentence.split()
unique_words = set(words_in_sentence)

print("Unique words count: ",len(unique_words))
print("Unique Words: ", unique_words)
```

# Highest Salary from Employee Data

```python
employees = {"Ravi": 75000, "Anita": 68000,"Kiran": 72000}

max_salary = 0
max_employee = ""

for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        max_employee = name

print(f"Highest Salary: {max_employee} - {max_salary}")
```

# To find maximum and minimum values

```python

numbers = [50, 32, 69, 90, 44]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)


# if you want output for the nmbers you entered

# numbers = input("Enter The Numbers:")
# number_list = numbers.split()
# max_num = numbers[0]
# min_num = numbers[0]
# for num in number_list :
#     if num > max_num:
#         max_num = num
#     if num < min_num:
#         min_num = num

# print("List of numbers which you Entered: ",number_list)
# print("Max Value:",max_num)
# print("Minimum Value: ",min_num)
```

# To count products above a price Threshold

```python
prices = input("Enter the prices :")
prices_list = prices.split()
count = 0
for price in prices_list:
    price = int(price)
    if price > 1000:
        count += 1

print("Products above 1000: ",count)
```

# Calculate attendance percentage

```python
attendance = input("Enter the attendance (P/A) : ")
attendance_list = attendance.split()
present = 0
for day in attendance:
    if day == "p" :
        present += 1

total_days = len(attendance)

percent_attendance = (present / total_days)* 100 
print("Attendance Percentage : ",percent_attendance)


#You just have to enter p or a in lower case
```

# Count Character Frequency

```python
characters = input("Enter a string: ")
frequency = {}
for char in characters:
    if char in frequency:
        frequency[char] += 1
    else :
        frequency[char] = 1

print(frequency)
```
